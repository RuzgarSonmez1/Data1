#4

Counting Lead Studios then visiulasing it

```python
def barhChart(col):
    raw = Counter(df1[col])
    labels= list(raw.keys())
    values=raw.values()
    plt.barh(labels, values)
    plt.show()

plt.rcParams['figure.figsize'] = 12,6
barhChart('Lead Studio')
```

```python
def pieChart(col, ex=None):
    raw = Counter(df1[col])
    ks = raw.keys()
    vs = 100*np.array(list(raw.values()))/sum(raw.values())
    plt.pie(vs, labels=ks, explode=ex)
    plt.show()
plt.rcParams['figure.figsize'] = 6,6
pieChart('Lead Studio')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat',
                  names=["Airline","Airline ID" ,"Source airport","Source airport ID", "Destination airport", "Coverage","Codeshare","Stops","Equipment"])
data.dropna()
```

#2

data contains flights informations such as airline,airpot(source and destination) 

#3

Construct a dataframe that includes 2 rows from routes data and add 1 more row that constat 1 to sum  or count flight between airports.

```python
d = {'source': data['Source airport'], 'target': data['Destination airport']}
df = pd.DataFrame(data=d)
df['nof']=1
df2=df.groupby(['source','target']).sum().sort_values(by=['nof'])
df2
```

#4

Filtering data above to find 10 or more flights between them

```python
df3=df2[df2.nof>=10]
df3.reset_index(inplace=True)
df3
```

#5

using previous dataframe I constrct to visiulize as network

```python
plt.rcParams['figure.figsize'] = 12,7

graph = nx.from_pandas_edgelist(df3)
nx.draw_kamada_kawai(graph,with_labels=True)
```
